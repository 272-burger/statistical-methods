#### 10. ì–´ëŠ ë²”ì£„í•™ ì—°êµ¬ìê°€ ì¸êµ¬ë°€ë„ì™€ ì ˆë„ë°œìƒë¥  ê°„ì˜ ê´€ê³„ë¥¼ ì—°êµ¬í•˜ë©´ì„œ ë‹¤ìŒì˜ 16ê°œ ë„ì‹œì˜ ìë£Œë¥¼ ìˆ˜ì§‘í•˜ì˜€ë‹¤. XëŠ” í•´ë‹¹ë„ì‹œì˜ ë‹¨ìœ„ë©´ì ë‹¹ ì¸êµ¬ë°€ë„ë¥¼, YëŠ” ì´ì „ë…„ë„ì˜ 10ë§Œëª… ë‹¹ ì ˆë„ë²”ì£„ì˜ ë°œìƒ íšŸìˆ˜ë¥¼ ì¡°ì‚¬í•œ ê²ƒì´ë‹¤. 

```r
# Assigning Data
population <- c(59, 49, 75, 54, 78,
                56, 60, 82, 69, 83, 
                88, 94, 47, 65, 89, 70)
crime <- c(209, 180, 195, 192, 215,
           197, 208, 189, 213, 201, 
           214, 212, 205, 186, 200, 204)
(q10_df <- data.frame(population, crime))
```

#### a. ìœ„ì˜ ë‘ ë³€ìˆ˜ ê°„ì˜ ê´€ê³„ë¥¼ ê°€ì§€ê³  íšŒê·€ì‹ì„ ì¶”ì •í•˜ì—¬ë¼. ì¶”ì •ëœ íšŒê·€ì‹ê³¼ ìë£Œë¥¼ ê°€ì§€ê³  ê·¸ë¦¼ì„ ê·¸ë ¤ ë‚˜íƒ€ë‚´ ë³´ê³ , ì„ í˜• íšŒê·€ë°©ì •ì‹ì´ ì˜ ë“¤ì–´ë§ëŠ”ì§€ì— ëŒ€í•´ ë§í•˜ì—¬ë¼. 

ì£¼ì–´ì§„ ìë£Œì—ì„œ ì„¤ëª…ë³€ìˆ˜ëŠ” í•´ë‹¹ë„ì‹œì˜ 'ë‹¨ìœ„ ë©´ì ë‹¹ ì¸êµ¬ë°€ë„' í•˜ë‚˜ì´ê³ , ë°˜ì‘ë³€ìˆ˜ëŠ” 'ì´ì „ë…„ë„ì˜ 10ë§Œëª… ë‹¹ ì ˆë„ë²”ì£„ì˜ ë°œìƒíšŸìˆ˜'ì´ë‹¤. ìš°ì„  ìµœì†Œì œê³±ì¶”ì •ë²•ì„ ì ìš©í•˜ì—¬ íšŒê·€ì‹ì„ ì¶”ì •í•˜ê³ ì í•œë‹¤. 

ìµœì†Œì œê³±ì¶”ì •ë²•ìœ¼ë¡œ êµ¬í•œ íšŒê·€ì„ ì´ ìœ íš¨í•˜ê¸° ìœ„í•´ì„  ë‹¤ìŒê³¼ ê°™ì€ ë‹¨ìˆœì„ í˜•íšŒê·€ëª¨í˜•ì˜ ê°€ì •ì‚¬í•­ë“¤ì„ ì¶©ì¡±í•´ì•¼ í•œë‹¤.

$$

1. ë“±ë¶„ì‚°ì„±: ëª¨ë“  ì˜¤ì°¨í•­( ğœ–ğ‘– )ë“¤ì˜ í‰ê· ì€ 0ì´ê³ , ë¶„ì‚°ì€  ğœ2 ë¡œ ë™ì¼í•˜ë‹¤.

3. ì •ê·œì„±: ì˜¤ì°¨í•­( ğœ–ğ‘– )ì€ ì •ê·œë¶„í¬ë¥¼ ë”°ë¥¸ë‹¤.

ê°€ì •ì‚¬í•­ì„ ì¶©ì¡±í•œë‹¤ëŠ” ê°€ì • í•˜ì— ìµœì†Œì œê³±ì¶”ì •ë²•ìœ¼ë¡œ yì ˆí¸ê³¼ ê¸°ìš¸ê¸°ë¥¼ ì¶”ì •í•  ìˆ˜ ìˆë‹¤. ìš°ì„  lmí•¨ìˆ˜ë¥¼ ì´ìš©í•˜ì—¬ êµ¬í•˜ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤.

```r
# Linear model
q10_fit <- lm(q10_df$crime ~ q10_df$population)
summary(q10_fit)
```

ì§ì ‘ ê³µì‹ì— ëŒ€ì…í•˜ì—¬ êµ¬í•œ ê°’ì€ ë‹¤ìŒê³¼ ê°™ë‹¤.

```r
# Least Squares Method

## b1
sd.x_q10 <- sd(q10_df$population)
sd.y_q10 <- sd(q10_df$crime)
(b1_q10 <- cor_q10 * sd.y_q10 / sd.x_q10)

## b0
mean.x_q10 <- mean(q10_df$population)
mean.y_q10 <- mean(q10_df$crime)
(b0_q10 <- mean.y_q10 - b1_q10 * mean.x_q10)
```

yì ˆí¸  ğ›½0 ê³¼ ê¸°ìš¸ê¸°  ğ›½1 ì˜ ì¶”ì •ê°’  ğ‘0 ,  ğ‘1 ëŠ” ê°ê° 182.9725, 0.2616ì´ë‹¤. ë”°ë¼ì„œ íšŒê·€ì‹ì€ ë‹¤ìŒê³¼ ê°™ì´ êµ¬í•´ì§„ë‹¤.

íšŒê·€ëª¨í˜•ì˜ ì í•©ì„±(goodness of fit)ì„ íŒë‹¨í•˜ê¸° ìœ ìš©í•œ ì²™ë„ëŠ” ê²°ì •ê³„ìˆ˜ì´ë‹¤. ê²°ì •ê³„ìˆ˜ëŠ” ìë£Œ ì „ì²´ì˜ ë³€ë™ëŸ‰ ì¤‘ íšŒê·€ì‹ìœ¼ë¡œ ì„¤ëª…í•  ìˆ˜ ìˆëŠ” ë³€ë™ëŸ‰ì˜ ë¹„ìœ¨ì„ ë‚˜íƒ€ë‚¸ ê°’ìœ¼ë¡œ, 1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ì¶”ì •ëœ íšŒê·€ì‹ì´ ì´ë³€ë™ëŸ‰ì˜ ë§ì€ ë¶€ë¶„ì„ ì„¤ëª…í•œë‹¤ê³  íŒë‹¨í•  ìˆ˜ ìˆë‹¤. ë”°ë¼ì„œ ê²°ì •ê³„ìˆ˜ë¥¼ êµ¬í•¨ìœ¼ë¡œì¨ ì¶”ì •í•œ íšŒê·€ì‹ì´ ì í•©í•œì§€ íŒë‹¨í•˜ê³ ì í•œë‹¤. 

```r
# Goodness of fit
y.hat_q10 <- b0_q10 + (b1_q10 * q10_df$population)

SSE_q10 <- sum((q10_df$crime - y.hat_q10)^2)
SSR_q10 <- sum((y.hat_q10 - mean.y_q10)^2)
SST_q10 <- SSE_q10 + SSR_q10
(r.squared_q10 <- SSR_q10 / SST_q10)
(cor_q10 ^ 2)
```

#### 12. ì–´ëŠ ê°€ìŠ¤ê³µê¸‰íšŒì‚¬ê°€ ì§€ë‚œ 1ë…„ê°„ ë§¤ì›” ì¡°ì‚¬í•œ ë‚œë°©ë„ì¼(heating degree-days, ê·œì •ë‚œë°©ì˜¨ë„ ì´í•˜ë¡œ ë–¨ì–´ì§„ ë‚ ì—, (ê·œì •ì˜¨ë„ - í•´ë‹¹ ì¼ì˜ ì˜¨ë„)í•œ ê²ƒì„ í•œ ë‹¬ ë™ì•ˆ ëˆ„ì í•œ ê°’)ê³¼ ë‚œë°©ìš© ê°€ìŠ¤ì†Œë¹„ëŸ‰(in hundreds of cubic meters)ì´ ë‹¤ìŒê³¼ ê°™ì´ ë‚˜íƒ€ë‚¬ë‹¤.

#### a. ìœ„ì˜ ìë£Œë¡œ ì–»ëŠ” ìµœì†Œì œê³± íšŒê·€ì„ ì€ ë¬´ì—‡ì¸ê°€?

ìµœì†Œì œê³±ì¶”ì •ë²•ìœ¼ë¡œ êµ¬í•œ íšŒê·€ì„ ì´ ìœ íš¨í•˜ê¸° ìœ„í•´ì„  ë‹¤ìŒê³¼ ê°™ì€ ë‹¨ìˆœì„ í˜•íšŒê·€ëª¨í˜•ì˜ ê°€ì •ì‚¬í•­ë“¤ì„ ì¶©ì¡±í•´ì•¼ í•œë‹¤.

ê°€ì •ì‚¬í•­ì„ ì¶©ì¡±í•œë‹¤ëŠ” ê°€ì • í•˜ì— ìµœì†Œì œê³±ì¶”ì •ë²•ìœ¼ë¡œ yì ˆí¸ê³¼ ê¸°ìš¸ê¸°ë¥¼ ì¶”ì •í•  ìˆ˜ ìˆë‹¤. ìš°ì„  lmí•¨ìˆ˜ë¥¼ ì´ìš©í•˜ì—¬ êµ¬í•˜ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤.

```r
# b1 Confidence Interval

t_gas <- qt(0.05, 7, lower.tail = F)
s.b1_gas <- sqrt(MSE_gas/sum((gas_df$heating_degree - mean.x_gas)^2))
(b1_CI.low_gas <- b1_gas - t_gas * s.b1_gas)
(b1_CI.up_gas <- b1_gas + t_gas * s.b1_gas)
```

ê¸°ìš¸ê¸° ì¶”ì •ì¹˜ì— ëŒ€í•œ 90% ìˆ˜ì¤€ì˜ ì‹ ë¢°êµ¬ê°„ì€ (0.18, 0.22)ì´ë©°, 0ì„ í¬í•¨í•˜ì§€ ì•ŠëŠ”ë‹¤. ë”°ë¼ì„œ ìœ ì˜ìˆ˜ì¤€ $\alpha = 0.1$í•˜ì—ì„œ ê·€ë¬´ê°€ì„¤ì€ ê¸°ê°ëœë‹¤. ì‹ ë¢°êµ¬ê°„ì„ í†µí•´ì„œë„ ì¶”ì •ëœ íšŒê·€ì‹ì´ ë‘ ë³€ìˆ˜ ê°„ ì„ í˜•ê´€ê³„ë¥¼ ì˜ ë‚˜íƒ€ë‚´ê³  ìˆë‹¤ê³  íŒë‹¨í•  ìˆ˜ ìˆë‹¤. 

#### ì¡°ì‚¬ìë£Œì—ì„œ 1ì›”ì˜ ê°€ìŠ¤ì‚¬ìš©ëŸ‰ ìˆ˜ì¹˜ê°€ 8.5ì—ì„œ 85ë¡œ ì˜ëª» ì…ë ¥ë˜ì—ˆë‹¤ê³ ê°€ì •í•˜ê³ , ì•„ë˜ ë¬¸ì œë“¤ì— ë‹µí•˜ì—¬ë¼.
#### e. ì˜¤ë¥˜ê°€ ìˆëŠ” 1ì›” ì¡°ì‚¬ ê´€ì°°ì¹˜ê°€ í¬í•¨ëœ ìë£Œë¥¼ ê°€ì§€ê³  íšŒê·€ë¶„ì„ì„ í•˜ê³ , ìµœì†Œì œê³± íšŒê·€ì„ ì„ ë‹¤ì‹œ ì¨ë³´ë¼.

```r
# Assigning data
month <- c("10", "11", "12", "1", "2", "3", "4", "5", "6")
heating_degree <- c(15.6, 26.8, 37.8, 36.4, 35.5, 18.6, 15.3, 7.9, 0.0)
gas_usage_error <- c(5.2, 6.1, 8.7, 85, 8.8, 4.9, 4.5, 2.5, 1.1)
gas_df_error <- data.frame(month, heating_degree, gas_usage_error)

# Scatter plot
plot(gas_df_error$gas_usage_error ~ gas_df_error$heating_degree)
cor_gas_error <- cor(gas_df_error$gas_usage_error, gas_df_error$heating_degree)
```

```r
# Least Squares Method
## b1
sd.x_gas <- sd(gas_df$heating_degree)
sd.y_gas_error <- sd(gas_df_error$gas_usage_error)
(b1_gas_error <- cor_gas_error * sd.y_gas_error / sd.x_gas)

## b0
mean.x_gas <- mean(gas_df_error$heating_degree)
mean.y_gas_error <- mean(gas_df_error$gas_usage_error)
(b0_gas_error <- mean.y_gas_error - b1_gas_error * mean.x_gas)
```

#### 13. ë‹¤ìŒì— ì£¼ì–´ì§„ 10ê°œ í‘œë³¸ê°€êµ¬ì˜ ì›”ì†Œë“ì•¡ê³¼ ì›” í†µì‹ ë¹„ì§€ì¶œì•¡ì˜ ìë£Œë¥¼ ì°¸ì¡°í•˜ì—¬ ë‹µì„ êµ¬í•˜ì—¬ë¼.

```r
# Assigning data
income <- c(160, 450, 360, 320, 300, 130, 410, 150, 360, 400)
expense <- c(35, 78, 102, 56, 75, 26, 130, 42, 59, 85)
(q13_df <- data.frame(income, expense))
```

#### a. ì›”í†µì‹ ë¹„ì§€ì¶œì•¡(y)ì— ëŒ€í•´ì„œ ì›”ì†Œë“ì•¡(x)ì„ ì„¤ëª…ë³€ìˆ˜ë¡œ í•˜ëŠ” ë‹¨ìˆœíšŒê·€ëª¨í˜•ë°©ì •ì‹ì„ ì„¤ì •í•˜ê³  ëª¨ìˆ«ê°’ì„ ì¶”ì •í•˜ì—¬ë¼. 

ìš°ì„  ë‘ ë³€ìˆ˜ì˜ ê´€ê³„ë¥¼ ì‚°ì ë„ë¡œ ë‚˜íƒ€ë‚´ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤.

```r
# Scatter plot
plot(q13_df$expense ~ q13_df$income)

# Correlation coefficient
(cor_q13 <- cor(q13_df$expense, q13_df$income))
```

ì‚°ì ë„ë¥¼ ì°¸ê³ í–ˆì„ ë•Œ, ì „ë°˜ì ìœ¼ë¡œ ë‘ ë³€ìˆ˜ ê°„ì—ëŠ” ì„ í˜•ê´€ê³„ê°€ ì¡´ì¬í•œë‹¤ê³  íŒë‹¨í•  ìˆ˜ ìˆë‹¤. ì‹¤ì œë¡œ ìƒê´€ê³„ìˆ˜ë„ 0.80ì´ë‹¤.

ì›”í†µì‹ ë¹„ì§€ì¶œì•¡(y)ì— ëŒ€í•´ì„œ ì›”ì†Œë“ì•¡(x)ì„ ì„¤ëª…ë³€ìˆ˜ë¡œ í•˜ëŠ” ë‹¨ìˆœíšŒê·€ëª¨í˜•ë°©ì •ì‹ì€ ë‹¤ìŒê³¼ ê°™ì´ ì„¤ì •í•  ìˆ˜ ìˆë‹¤. 

> $Y= \beta_{0} + \beta_{1}X + \epsilon$

ìµœì†Œì œê³±ì¶”ì •ë²•ì„ ì ìš©í•˜ì—¬ ëª¨ìˆ˜ $\beta_{0}$ì™€ $\beta_{1}$ì„ ì¶”ì •í•˜ê³ ì í•œë‹¤. ìµœì†Œì œê³±ì¶”ì •ë²•ìœ¼ë¡œ êµ¬í•œ íšŒê·€ì„ ì´ ìœ íš¨í•˜ê¸° ìœ„í•´ì„  ë‹¤ìŒê³¼ ê°™ì€ ë‹¨ìˆœì„ í˜•íšŒê·€ëª¨í˜•ì˜ ê°€ì •ì‚¬í•­ë“¤ì„ ì¶©ì¡±í•´ì•¼ í•œë‹¤. 
 
1. ë“±ë¶„ì‚°ì„±: ëª¨ë“  ì˜¤ì°¨í•­( ğœ–ğ‘– )ë“¤ì˜ í‰ê· ì€ 0ì´ê³ , ë¶„ì‚°ì€  ğœ2 ë¡œ ë™ì¼í•˜ë‹¤.
2. ë…ë¦½ì„±: ì˜¤ì°¨í•­( ğœ–ğ‘– )ë“¤ì€ ì„œë¡œ ë…ë¦½ì´ë‹¤.
3. ì •ê·œì„±: ì˜¤ì°¨í•­( ğœ–ğ‘– )ì€ ì •ê·œë¶„í¬ë¥¼ ë”°ë¥¸ë‹¤.

ê°€ì •ì‚¬í•­ì„ ì¶©ì¡±í•œë‹¤ëŠ” ê°€ì • í•˜ì— ìµœì†Œì œê³±ì¶”ì •ë²•ìœ¼ë¡œ yì ˆí¸ê³¼ ê¸°ìš¸ê¸°ë¥¼ ì¶”ì •í•  ìˆ˜ ìˆë‹¤. ìš°ì„  lmí•¨ìˆ˜ë¥¼ ì´ìš©í•˜ì—¬ êµ¬í•˜ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤.

```r
# Linear model
q13_fit <- lm(q13_df$expense ~ q13_df$income)
summary(q13_fit)
```

ì§ì ‘ ê³µì‹ì— ëŒ€ì…í•˜ì—¬ êµ¬í•˜ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤. 

```r
# Least Squares Method
## b1
sd.x_q13 <- sd(q13_df$income)
sd.y_q13 <- sd(q13_df$expense)
(b1_q13 <- cor_q13 * sd.y_q13 / sd.x_q13)

## b2
mean.x_q13 <- mean(q13_df$income)
mean.y_q13 <- mean(q13_df$expense)
(b0_q13 <- mean.y_q13 - b1_q13 * mean.x_q13)
```

yì ˆí¸ $\beta_{0}$ê³¼ ê¸°ìš¸ê¸° $\beta_{1}$ì˜ ì¶”ì •ê°’ $b_{0}$, $b_{1}$ëŠ” ê°ê° 0.2187, 2.3173ì´ë‹¤. ë”°ë¼ì„œ íšŒê·€ì‹ì€ ë‹¤ìŒê³¼ ê°™ì´ êµ¬í•´ì§„ë‹¤.
$$
\hat{y} = 0.2187 + 2.3173x
$$

ì¶”ì •ëœ íšŒê·€ì‹ê³¼ ìë£Œë“¤ì„ ê·¸ë¦¼ìœ¼ë¡œ ë‚˜íƒ€ë‚´ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤. 

```r
# Visualization
plot(q13_df$expense ~ q13_df$income)
abline(q13_fit)
```

#### b. $SS_{xx}$,  $SS_{yy}$,  $SS_{xy}$ë¥¼ êµ¬í•˜ì—¬ë¼.
ì°¨ë¡€ëŒ€ë¡œ 123040, 9185.6, 26908ì´ë‹¤. 


```r
# Correlation coefficient
(ss_xx <- sum((q13_df$income - mean(q13_df$income))^2))
(ss_yy <- sum((q13_df$expense - mean(q13_df$expense))^2))
(ss_xy <- sum((q13_df$income - mean(q13_df$income))*(q13_df$expense - mean(q13_df$expense))))
```

#### c. ì˜¤ì°¨í•­ì˜ í‘œì¤€í¸ì°¨ë¥¼ êµ¬í•˜ì—¬ë¼. 
ì˜¤ì°¨í•­ì˜ ë¶„ì‚°($\hat{\sigma}^2$)ì€ ì˜¤ì°¨í‰ê· ì œê³±(MSE)ìœ¼ë¡œ ì¶”ì •ëœë‹¤. ì´ë•Œ ììœ ë„ëŠ” 8(= 10 - 2)ì´ë‹¤. 

```r
y.hat_q13 <- b0_q13 + (b1_q13 * q13_df$income)
SSE_q13 <- sum((q13_df$expense - y.hat_q13)^2)
MSE_q13 <- SSE_q13/(10 - 2)

(sb1_q13 <- sqrt(MSE_q13/8))
```

#### d. SST, SSE, SSRë¥¼ êµ¬í•˜ì—¬ë¼.
ìˆœì„œëŒ€ë¡œ 3301.01, 5884.59, 9185.6ì´ë‹¤. 

```r
# Sum of Squares

(SSE_q13 <- sum((q13_df$expense - y.hat_q13)^2))
(SSR_q13 <- sum((y.hat_q13 - mean.y_q13)^2))
(SST_q13 <- SSE_q13 + SSR_q13)
```

#### e. ì´ ì„ í˜•ëª¨í˜•ì„ ê°€ì§€ê³  ì–»ëŠ” ê²°ì •ê³„ìˆ˜ëŠ” ì–¼ë§ˆì¸ê°€? 
íšŒê·€ì„ ì´ ìë£Œë¥¼ ì˜ ë°˜ì˜, ìš”ì•½í•˜ê³  ìˆëŠ”ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì²™ë„ê°€ ê²°ì •ê³„ìˆ˜ë‹¤. ê²°ì •ê³„ìˆ˜ëŠ” ìë£Œ ì „ì²´ì˜ ë³€ë™ëŸ‰ ì¤‘ íšŒê·€ì‹ìœ¼ë¡œ ì„¤ëª…í•  ìˆ˜ ìˆëŠ” ë³€ë™ëŸ‰ì˜ ë¹„ìœ¨ì„ ë‚˜íƒ€ë‚¸ ê°’ìœ¼ë¡œ, êµ¬í•˜ëŠ” ì‹ì€ ë‹¤ìŒê³¼ ê°™ë‹¤.

```r
# R squared 
(SSR_q13/SST_q13)
```
